```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import recall_score, precision_score
from sklearn.model_selection import cross_validate
from sklearn.model_selection import GridSearchCV
```

```python
# Read in our data set
cancer = pd.read_csv('dataset/wdbc.csv')
cancer
```

```python
cancer["diagnosis"] = cancer["diagnosis"].replace({
    "M" : "Malignant",
    "B" : "Benign"
})
```

```python
cancer
```

```python
statardized_cancer = cancer.copy()
column_to_exclude = ['id','diagnosis']
column_to_scale = statardized_cancer.columns.difference(column_to_exclude)
```

```python
scaler = StandardScaler()
statardized_cancer[column_to_scale] = scaler.fit_transform(statardized_cancer[column_to_scale])
```

```python
statardized_cancer
```

```python
# set the seed
np.random.seed(1)

# Split the data
cancer_train, cancer_test = train_test_split(statardized_cancer, train_size=0.75, shuffle = True, stratify = statardized_cancer["diagnosis"])
```

```python
cancer_train.info()
```

```python
cancer_test.info()
```

```python
# k = 5

knn = KNeighborsClassifier(n_neighbors = 5)
```

```python
# step 2: define our X and y

X = cancer_train[["perimeter_mean", "concavity_mean"]]
y = cancer_train["diagnosis"]
```

```python
knn.fit(X,y)
```

```python
cancer_test["predicted"] = knn.predict(cancer_test[["perimeter_mean","concavity_mean"]])
```

```python
cancer_test
```

```python
knn.score(cancer_test[["perimeter_mean", "concavity_mean"]], cancer_test["diagnosis"])
```

```python
# confusion matrix

pd.crosstab(
    cancer_test["diagnosis"],
    cancer_test["predicted"]
)
```

```python
precision_score(
    y_true = cancer_test["diagnosis"], 
    y_pred = cancer_test["predicted"],
    pos_label = "Malignant"
)
```

```python
recall_score(
    y_true = cancer_test["diagnosis"], 
    y_pred = cancer_test["predicted"],
    pos_label = "Malignant"
)
```
